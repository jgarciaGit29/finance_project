```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# EDA: Exploratory Data Analysis
tesla_stock_historical = r'../data/cleaned/Tesla_Historial_data_clean.csv'
df_tesla_stock_historical = pd.read_csv(tesla_stock_historical)
```

```python
#plot histogram for Open, High, Low, Close, prices
# This will help you understand the distribution of Tesla Historical Stock data.
df_tesla_stock_historical[['Open','High','Low','Close']].hist(bins=50,figsize=(12,10))
                                                              
plt.suptitle('Distribution of Telsa Stock Prices')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import pytz

# Load the data and ensure 'Date' column is included
df_tesla = pd.read_csv(r'../data/cleaned/Tesla_Historial_data_clean.csv')
print("Columns:", df_tesla.columns)  # Verify column names

# Step 1: Convert 'Date' to datetime with utc=True and set as index in NYC zone
df_tesla['Date'] = pd.to_datetime(df_tesla['Date'], utc=True)
df_tesla['Date'] = df_tesla['Date'].dt.tz_convert('America/New_York')
df_tesla.set_index('Date', inplace=True)

# Step 2: Plot the time series for Open, High, Low, Closee pricess
plt.figure(figsize=(14, 7))  # Set figure size

# Plot each price series
plt.plot(df_tesla.index, df_tesla['Open'], label='Open')
plt.plot(df_tesla.index, df_tesla['High'], label='High')
plt.plot(df_tesla.index, df_tesla['Low'], label='Low')
plt.plot(df_tesla.index, df_tesla['Close'], label='Close')

# Step 3: Format the date labels on the x axis
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Major ticks at start of each year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format x-axis labels as year

# Add labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Time Series of Tesla Stock Prices')
plt.legend()

# Rotate date labels for readability
plt.xticks(rotation=45)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
